```python
def download(url):
    print 'Downloading:',url
    try:
        html = urllib2.urlopen(url).read().decode('utf-8')
    except urllib2.URLError as e:
        print 'Download Error:',e.reason
        html = None
    return html
```

```python
def getTagByClass(bsobj,tagName,classValue):
#     print 'Get %s by class:%s'%(tagName,classValue)
    try:
        value = bsobj.findAll(tagName,{"class":classValue})
    except AttributeError as e:
        return None
    return value
```

```python
def runCrawler(catalog):
    # 生成要抓取的list地址 1000篇文章
    for i in range(1,41):
        url = 'https://news.yahoo.co.jp/hl?c=' + catalog + '&p=' + str(i)
        html = download(url)
        time.sleep(3)
        soup = BeautifulSoup(html,"lxml")
        # 获取预测的文章的列表
        forecastList = getTagByClass(soup,"p","ttl")
        # 初始化预测页面集合
        pages = set()
        # 获取每篇预测的URL
        for item in forecastList:
            link = item.find("a",href=re.compile("^https://headlines.yahoo.co.jp/hl*"))
            # 経済総合 新着ニュース 以外的新闻不做抓取
            if link is not None:
    #             print item.text
                newpage =  link.attrs['href']
                # 如果没有在预测页面集合里面则加入
                if newpage not in pages:
    #                 print newpage
                    # 加入预测页面集合
                    pages.add(newpage)

        urlTarget = set()
        # 生成完整地址
        for pageUrl in pages:
    #         print pageUrl
            urlTarget.add(pageUrl)   

        #获取预测文章内容
        for target in urlTarget:

            # 文件名
            csvFile = target.split("=")[-1]

            # 获取网页
            html = download(target)
            time.sleep(1)
            bsobj = BeautifulSoup(html,"lxml")

            title = ""
            title = bsobj.findAll("h1")[0].get_text().encode('utf-8') 

            # 预测内容 getTagByClass(bsobj,"p","ynDetailText")
            yjm_brandExp = getTagByClass(bsobj,"p","ynDetailText")
            content = ""

            for item in yjm_brandExp:        
                content = content + item.get_text().encode('utf-8') 

            # 写入CSV
            # 需要预先在程序相同的目录下创建files文件夹
            output = open("/home/hadoop/Crawler/" + catalog + "/" + csvFile + ".csv",'w+')
            fields = ('title','content')
            try:
                writer = csv.writer(output)
                data = (title,content)
                writer.writerow(fields)
                writer.writerow(data)
            finally:
                output.close()
            print "CSV OUTPUT:",csvFile
```
